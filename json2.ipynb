```python
import json  
from pprint import pprint 
with open('recipes.json') as f:
    recipes = json.load(f)
```

```python
import pandas as pd
df = pd.DataFrame(recipes)
```

```python
df.head()
```

```python
def find_item(cell):
    if item in cell:
        return 1
    return 0
```

```python
ingredients = list(df.columns[3:])
```

```python
for item in ingredients:
    df[item] = df['ingredients'].apply(find_item)
```

```python
df['ingredients'] = df['ingredients'].apply(lambda x: len(x))
```

```python
df.to_csv('recipes.csv', index = False)
```

```python
df = pd.read_csv('recipes.csv')
```

```python
ids = pd.Series(df.id.values).tolist()
ingredients = df.columns[3:].tolist()
new_recipes = []
for current_id in ids:
    recipe_id = int(current_id)
    recipe_cuisine = df[df['id'] == recipe_id]['cuisine'].iloc[0]
    recipe_ingredients = df[df['id'] == recipe_id].columns[df[df['id'] == recipe_id].apply(lambda x: True if x.iloc[0]==1 else False, axis=0)].tolist()
    recipe = {'cuisine': recipe_cuisine, 'id': recipe_id, 'ingredients': recipe_ingredients}
    new_recipes.append(recipe)
```
